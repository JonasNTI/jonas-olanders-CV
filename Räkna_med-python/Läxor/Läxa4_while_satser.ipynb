```python
# Uppgift 1

s = 0
i = 0

while i < 100:
    i += 1
    s += i
print(s)
```

```python
# Uppgift 3

s = 0
n = 0

while n <= 5:
    s = 2 ** -n
    n = n + 1

print(s)
```

```python
# Uppgift 5

import random as rnd
n = rnd.randint(0,10)

Gissning = int(input("Gissa ett tal mellan 0 och 100"))

while Gissning != n:
    if Gissning < n:
        Gissning = int(input("Större"))
    elif Gissning > n:
        Gissning = int(input("Mindre"))

print(f"Du svarade rätt! Talet var {Gissning}")
        

#while Gissning < n:
#    Gissning = int(input("Större"))
    
#while Gissning > n:
#    Gissning = int(input("Mindre"))

# print(f"Du svarade rätt! Talet var {Gissning}")
```
